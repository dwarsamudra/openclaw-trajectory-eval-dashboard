```python
import json
import pandas as pd
from pathlib import Path

# Paths
trajectory_folder = Path("F:/openclaw-trajectory-eval-lab/data/trajectories")
scores_path = Path("F:/openclaw-trajectory-eval-lab/evaluation/trajectory_scores.csv")
evidence_path = Path("F:/openclaw-trajectory-eval-lab/data/evidence/evidence_tracker.csv")

# Load trajectory JSONs
files = sorted(trajectory_folder.glob("task_*_trajectory.json"))
print("FILES FOUND:", [f.name for f in files])

rows = []
for file in files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
        rows.append({
            "task_id": int(data["task_id"]),
            "task_title": data["task_title"],
            "difficulty": data["difficulty"],
            "status": data["status"],
            "trajectory_score": float(data["review"]["overall_score"])
        })

df = pd.DataFrame(rows)

# Load CSVs
scores_df = pd.read_csv(scores_path)
evidence_df = pd.read_csv(evidence_path)

# Clean column names
df.columns = [str(c).strip() for c in df.columns]
scores_df.columns = [str(c).strip() for c in scores_df.columns]
evidence_df.columns = [str(c).strip() for c in evidence_df.columns]

# Force correct task_id types
df["task_id"] = pd.to_numeric(df["task_id"], errors="coerce").astype("Int64")
scores_df["task_id"] = pd.to_numeric(scores_df["task_id"], errors="coerce").astype("Int64")
evidence_df["task_id"] = pd.to_numeric(evidence_df["task_id"], errors="coerce").astype("Int64")

# Drop bad rows if any
df = df.dropna(subset=["task_id"]).copy()
scores_df = scores_df.dropna(subset=["task_id"]).copy()
evidence_df = evidence_df.dropna(subset=["task_id"]).copy()

df["task_id"] = df["task_id"].astype(int)
scores_df["task_id"] = scores_df["task_id"].astype(int)
evidence_df["task_id"] = evidence_df["task_id"].astype(int)

# Build evidence count
evidence_count = (
    evidence_df.groupby("task_id")
    .size()
    .reset_index(name="evidence_count")
)

# Merge safely
summary_df = (
    df.merge(
        scores_df[["task_id", "overall_score"]],
        on="task_id",
        how="left"
    )
    .merge(
        evidence_count,
        on="task_id",
        how="left"
    )
    .rename(columns={"overall_score": "review_score"})
)

print("\nDF COLUMNS:", df.columns.tolist())
print("SCORES COLUMNS:", scores_df.columns.tolist())
print("EVIDENCE COLUMNS:", evidence_df.columns.tolist())
print("EVIDENCE_COUNT COLUMNS:", evidence_count.columns.tolist())

print("\nSUMMARY:")
print(summary_df)
```

```python
import json
import pandas as pd
from pathlib import Path

trajectory_folder = Path("F:/openclaw-trajectory-eval-lab/data/trajectories")
files = sorted(trajectory_folder.glob("task_*_trajectory.json"))

rows = []
for file in files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
        rows.append({
            "task_id": int(data["task_id"]),
            "task_title": data["task_title"],
            "difficulty": data["difficulty"],
            "status": data["status"],
            "trajectory_score": float(data["review"]["overall_score"])
        })

df = pd.DataFrame(rows)
print(df)
print(df.columns.tolist())
```

```python
import pandas as pd

# Load CSVs
scores_df = pd.read_csv("F:/openclaw-trajectory-eval-lab/evaluation/trajectory_scores.csv")
evidence_df = pd.read_csv("F:/openclaw-trajectory-eval-lab/data/evidence/evidence_tracker.csv")

# Clean column names
scores_df.columns = [str(c).strip() for c in scores_df.columns]
evidence_df.columns = [str(c).strip() for c in evidence_df.columns]

# Convert types
scores_df["task_id"] = scores_df["task_id"].astype(int)
evidence_df["task_id"] = evidence_df["task_id"].astype(int)

# Evidence count
evidence_count = evidence_df.groupby("task_id").size().reset_index(name="evidence_count")

# Merge everything
summary_df = df.merge(
    scores_df[["task_id", "overall_score"]],
    on="task_id",
    how="left"
).merge(
    evidence_count,
    on="task_id",
    how="left"
)

# Rename for clarity
summary_df = summary_df.rename(columns={"overall_score": "review_score"})

summary_df
```
